```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sb
from sklearn.cluster import (KMeans)

%matplotlib inline
```

```python
# Carrega os dados de um arquivo CSV
df = pd.read_csv('./dados/iris.csv')

# Remove a coluna com as classes
data = df.drop('target', axis=1)

# X recebe os dados limpos para o treinamento
X = np.array(data)

# define o número de clusters
kmeans = KMeans(n_clusters=3, random_state=0)
```

```python
# Treina os dados
kmeans.fit(X)
```

```python
# Exibe um array com os rótulos de cada item
kmeans.labels_
```

```python
# Adiciona uma coluna no dataframe com os novos cluters
data['labels'] = kmeans.labels_
```

```python
# Gera o gráfico com os grupos originais
sb.pairplot(df, hue='target')
```

```python
#Gera o gráfico com os grupos identificados pelo Kmeans
sb.pairplot(data, hue='labels')
```
