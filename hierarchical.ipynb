```python
import matplotlib.pyplot as plt  
import pandas as pd  
import numpy as np
import seaborn as sb
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
%matplotlib inline
```

```python
# Carrega os dados de um arquivo CSV
df = pd.read_csv('./dados/iris.csv')

# Remove a coluna com as classes
data = df.drop('target', axis=1)
```

```python
# Plota um dendograma dos dados
plt.figure(figsize=(8, 8))
plt.title("Dendograma")
plt.axhline(y=65)
dend = shc.dendrogram(shc.linkage(data, method='complete'))
```

```python
# Define os parâmetros do algoritmo de cluster
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')  
# Treina os dados
cluster.fit(data)
```

```python
# Adiciona uma coluna no dataframe com os novos cluters
data['labels'] = cluster.labels_

# Gera o gráfico com os grupos originais
sb.pairplot(df, hue='target')
```

```python
# Plota o gráfico com os grupos gerados pelo AgglomerativeClustering
sb.pairplot(data, hue='labels')
```
